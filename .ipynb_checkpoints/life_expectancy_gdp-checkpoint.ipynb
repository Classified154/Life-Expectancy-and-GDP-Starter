```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
data = pd.read_csv("all_data.csv")
```

```python
data.head()
```

```python
data = data.rename(columns={"Country":"country", "Year":"year", "Life expectancy at birth (years)": 'life_expec', "GDP": 'gdp'})
```

```python
data.info()
```

```python
plt.scatter(data.life_expec, data.country)
plt.show()
plt.clf()
```

```python
data.country.unique()
```

```python
sns.barplot(x="life_expec", y="country", data=data)
plt.show()
plt.clf()
```

```python
sns.lineplot(x=data.year, y=data.life_expec, hue=data.country)
plt.show()
plt.clf()
```

```python
sns.lineplot(x=data.year, y=data.gdp, hue=data.country)
plt.show()
plt.clf()
```

```python
con_z = data[data.country == "Zimbabwe"]
#con_z = normalise_gdp(con_z)

con_chile = data[data.country == "Chile"]
#con_chile = normalise_gdp(con_chile)

con_china = data[data.country == "China"]
#con_china = normalise_gdp(con_china)

con_ger = data[data.country == "Germany"]
#con_ger = normalise_gdp(con_ger)

con_mex = data[data.country == "Mexico"]
#con_mex = normalise_gdp(con_mex)

con_usa = data[data.country == 'United States of America']
#con_usa = normalise_gdp(con_usa)
```

```python
print("Average Life (Zimbabwe)", con_z.life_expec.mean())
print("Average Life (Chile)", con_chile.life_expec.mean())
print("Average Life (China)", con_china.life_expec.mean())
print("Average Life (Germany)", con_ger.life_expec.mean())
print("Average Life (Mexico)", con_mex.life_expec.mean())
print("Average Life (USA)", con_usa.life_expec.mean())
print("\nAverage Life (ALL)", data.life_expec.mean())
```

```python

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(12, 8), layout='constrained')

ax1.scatter(con_z.gdp, con_z.life_expec, s=10, c="green")
ax1.set_title("Zimbabwe")
ax1.set_xlabel("GPD (in Trillions)")
ax1.set_ylabel("Life Expectancy (years)")


ax2.scatter(con_chile.gdp, con_chile.life_expec, s=10, c="red")
ax2.set_title("Chile")
ax2.set_xlabel("GPD (in Trillions)")
ax2.set_ylabel("Life Expectancy (years)")

ax3.scatter(con_china.gdp, con_china.life_expec, s=10, c="blue")
ax3.set_title("China")
ax3.set_xlabel("GPD (in Trillions)")
ax3.set_ylabel("Life Expectancy (years)")

ax4.scatter(con_ger.gdp, con_ger.life_expec, s=10, c="orange")
ax4.set_title("Germany")
ax4.set_xlabel("GPD (in Trillions)")
ax4.set_ylabel("Life Expectancy (years)")

ax5.scatter(con_mex.gdp, con_mex.life_expec, s=10, c="purple")
ax5.set_title("Mexico")
ax5.set_xlabel("GPD (in Trillions)")
ax5.set_ylabel("Life Expectancy (years)")

ax6.scatter(con_usa.gdp, con_usa.life_expec, s=10, c="pink")
ax6.set_title("USA")
ax6.set_xlabel("GPD (in Trillions)")
ax6.set_ylabel("Life Expectancy (years)")

fig.legend(['Zimbabwe', 'Chile', 'China', 'Germany', 'Mexico', 'United States of America'], loc='outside upper right')

```
